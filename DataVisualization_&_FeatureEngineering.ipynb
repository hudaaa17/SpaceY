### TASK 1: Visualize the relationship between Flight Number and Launch Site




Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
sns.catplot(y = "LaunchSite", x = "FlightNumber", data = df, aspect = 2, height = 4, hue = "Class")
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.xticks(rotation = 45)
plt.show()
```

Findings:
- VAFB SLC 4E has a higher success rate overall
- At CCAFS SLC, as flight numbers increased no of successful landings also increased. This implies, reliability improved over time
- KSC LC also has high success rates


### TASK 2: Visualize the relationship between Payload Mass and Launch Site


```python
sns.catplot(y = "LaunchSite", x = "PayloadMass", data = df, hue = "Class", aspect = 2)
plt.xlabel("Pay Load Mass (kg)")
plt.ylabel("Launch Site")
plt.show()
```

- No site shows a perfect correlation between payload mass and success.

- KSC LC 39A seems to perform better for heavy payloads (near 16,000 kg) → most of these launches succeeded.

- CCAFS SLC 40 has the largest number of launches, but success/failure appears mixed at most payload levels.

- VAFB SLC 4E had fewer launches, so the pattern is less clear, but outcomes are mixed too.

### TASK  3: Visualize the relationship between success rate of each orbit type


mean → success rate (best for probability) ✅

sum → total successes

count → total launches

median/min/max → biased info, usually not helpful here

```python
orbit_success = df.groupby('Orbit')['Class'].mean()
orbit_success.plot(kind= "bar", color="skyblue", edgecolor="black")
plt.xlabel("Orbit")
plt.ylabel("Success Rate")
plt.title("Success Rate for Each Orbit")
plt.xticks(rotation=45)
plt.show()
```

GTO has the lowest success rate

GEO, SSO, ES-L1 have the highest success rates

    Implications:

    Missions targeting GEO/SSO/ES-L1 are relatively more reliable.

    GTO launches are riskier compared to others.

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


```python
sns.catplot(y = "Orbit", x = "FlightNumber", data = df, hue = "Class", aspect = 1.5)
plt.xlabel("Flight Number")
plt.xticks(rotation = 45)
plt.ylabel("Orbit Type")
plt.show()
```

With increasing Flights, the success rate of LEO has increased

Suggests GTO missions are technically challenging compared to LEO/ISS

Launches to ISS are generally more successful (mostly orange dots) across different flight numbers

Other orbits - Mostly successful but data points are fewer, so reliability can’t be judged strongly.


### TASK 5: Visualize the relationship between Payload Mass and Orbit type

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
```

```python
avg_success_rate = df.groupby("Year")["Class"].mean()
avg_success_rate.plot(kind="line")
plt.title("Average Launch Success Rate per Year")
plt.xlabel("Year")
plt.ylabel("Average Success Rate")
plt.show()
```

### Task 7: Create dummy variables for categorical columns
